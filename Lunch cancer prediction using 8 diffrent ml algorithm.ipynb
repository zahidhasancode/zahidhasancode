IMPORT LIBRARY 

```python
import numpy as np
import pandas as pd
import os 
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import FunctionTransformer
```

IMPORT ML ALGORITHM 

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
```

```python
df= pd.read_csv("/content/cancer patient data sets.csv")
```

```python
df.head(10)
```

```python
df.info()
```

```python
df.drop(["Patient Id" , "index"], axis = 1 , inplace = True)
```

```python
df
```

```python
df.replace(to_replace = "Low" , value = 0 , inplace = True)
df.replace(to_replace = "Medium" , value = 1 , inplace = True)
df.replace(to_replace = "High" , value = 2 , inplace = True)
```

```python
df
```

```python
a = df.drop("Level" , axis = 1)
b = df["Level"]
```

Function transformer 

```python
right_skew = []
left_skew = []
for i in df.columns:
    if df[i].skew() > 0:
        right_skew.append(i)
    else:
        left_skew.append(i)
```

```python
right_trf = FunctionTransformer(func = np.square)
left_trf = FunctionTransformer(func = np.log1p)
right_trfd = right_trf.fit_transform(df[right_skew])
left_trfd = left_trf.fit_transform(df[left_skew])

data_proc = pd.concat([right_trfd , left_trfd , b] , axis = 1 , join = "inner")
```

train and test data

```python
train , test = np.split(df.sample(frac = 1) , [int(0.8 * len(df))])
```

```python
def pre(dataframe):
    target = ["Level"]
    x = dataframe.drop(target , axis = 1)
    y = dataframe[target]
    
    return x , y
```

```python
X_train , Y_train = pre(train)
X_test , Y_test = pre(test)
```

Applying ml model 

KNeighborsRegressor

```python
model_0 = KNeighborsRegressor()
model_0.fit(X_train , Y_train)
```

LinearRegression

```python
model_1 = LinearRegression()
model_1.fit(X_train , Y_train)
```

 LogisticRegression

```python
model_2 = LogisticRegression()
model_2.fit(X_train , Y_train)
```

RidgeRegression

```python
model_3 = Ridge()
model_3.fit(X_train , Y_train)
```

Lasso Rigrassion 

```python
model_4 = Lasso()
model_4.fit(X_train , Y_train)
```

SupportVectorMachine

```python
model_5 = SVC()
model_5.fit(X_train , Y_train)
```

GaussianNaiveBayes

```python
model_6 = GaussianNB()
model_6.fit(X_train , Y_train)
```

RandomForestClassfier

```python
model_7 = RandomForestClassifier()
model_7.fit(X_train , Y_train)
```

```python
from sklearn.metrics import r2_score
```

```python
print("R2 score for " , model_0 , " is : " , r2_score(Y_test , model_0.predict(X_test)))
print("R2 score for " , model_1 , " is : " , r2_score(Y_test , model_1.predict(X_test)))
print("R2 score for " , model_2 , " is : " , r2_score(Y_test , model_2.predict(X_test)))
print("R2 score for " , model_3 , " is : " , r2_score(Y_test , model_3.predict(X_test)))
print("R2 score for " , model_4 , " is : " , r2_score(Y_test , model_4.predict(X_test)))
print("R2 score for " , model_5 , " is : " , r2_score(Y_test , model_5.predict(X_test)))
print("R2 score for " , model_6 , " is : " , r2_score(Y_test , model_6.predict(X_test)))
print("R2 score for " , model_7 , " is : " , r2_score(Y_test , model_7.predict(X_test)))
```
